# Extract Name,Rating and Review from all the review pages of the product

```python
cookie={}
reviews=[]
names=[]
rating=[]
for k in range(1,388):
        nList=[]
        rList=[]
        url="https://www.amazon.in/inches-Ready-UltraAndroid-32GA-Black/product-reviews/B07XMD275Q/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews"+'&pageNumber='+str(k)
        response=requests.get(url,cookies=cookie,headers=header)
        soup=BeautifulSoup(response.content)
        r=soup.findAll("span",{'data-hook':"review-body"})
        for i in r:
            reviews.append(i.text)
        n=soup.findAll("span","a-profile-name")
        for i in n:
            nList.append(i.text)
        nList.pop(0) #Deleting top review name with same class name
        nList.pop(0) #Deleting top review name with same class name
        names.extend(nList)
        ra=soup.findAll("span","a-icon-alt")
        for i in ra:
            rList.append(i.text)
        rList.pop(0) #Deleting the average rating given above the page with same class name
        rList.pop(0) #Deleting the top rating given above the page with same class name
        rList.pop(0) #Deleting the top rating given above the page with same class name
        rating.extend(rList)
        #print("k,rating",k,rating)
        print(url)
```

# Check the lengths of name,review and rating list and Convert all the lists to dictioneries(key:value pair)

```python
print(len(reviews))
print(len(names))
print(len(rating))
rev={'REVIEWS':reviews}
na={'NAMES':names}
rat={'RATING':rating}
```

# Create dataframes from dictionaries and concat all

```python
name=pd.DataFrame.from_dict(na)
reva=pd.DataFrame.from_dict(rev)
rati=pd.DataFrame.from_dict(rat)
review_data= pd.concat([name, reva],axis=1)
review_data= pd.concat([review_data, rati],axis=1)
```
